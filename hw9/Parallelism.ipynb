```python
from random import uniform
from math import sqrt
from time import time
from multiprocessing import Pool, cpu_count
from IPython.display import display
from IPython import parallel
import numpy as np
import os,sys
import pandas as pd
import scipy as sp
```

# Helper Functions

```python
def dart_throw(number_of_darts):
    '''
    simulates dart throwing. For each dart, find a random position in the unit square for it to fall. Test if it falls
    within the circle by calculating the distance from the origin (0.5, 0.5) to the dart. Darts that fall within 0.5
    of the origin are within the circle.
    
    input:
    number_of_darts -- number of darts thrown
    
    output:
    number_of_darts_in_circle -- the number of darts that landed in the quarter of the circle
    '''
    from random import uniform
    from math import sqrt
    
    #Define a variable to store the number of darts that fall inside the circle.
    number_of_darts_in_circle = 0
    
    for n in range(number_of_darts):
        x, y = uniform(0, 1), uniform(0, 1)
        if sqrt((x - 0.5)**2 + (y - 0.5)**2) <= 0.5:
            number_of_darts_in_circle += 1
    #record the time after the conculsion of the loop
    #end_time = time()
    
    #The total time required to run the loop is the difference.
    #execution_time = end_time - start_time
    
    return number_of_darts_in_circle #, execution_time

def run_simulation(num_of_darts, run_num = 0):
    '''
    runs dart throwing simulation in a serial fashion. The simulation calculates the execution time and darts per second thrown across the simulation length
    Simulation lenghth is determined by elements in the num_of_darts array, each entry specifies the length of the current simulation.
    input:
    number_of_darts -- an int array that specifices the number of darts to throw during a simulation.
    run_num -- if running multiple simulations of the same number of dart throws tags the trial number. 
    output:
    simulation_df -- returns a dataframe containing information for each simulation including the number of darts thrown in the simulation (Total_Darts), 
    the trial number (this is if you want to run through multiple simulations of the same length e.g. 5 simulations for 10, 100, 1000 etc darts thrown),
    the number of darts that landed in the circle (Darts_in_Circ), time to run the simulation of X number of dart throws (Execution_Time), and the 
    number of darts thrown per second during that simulation 
    '''
    simulation_df = pd.DataFrame(columns = ['Total_Darts', 'Trial', run_num, 'Darts_in_Circ', 'Execution_Time', 'Darts_per_second'], index = num_of_darts.astype(int))
    for dart_num in num_of_darts:
        start_time = time()
        in_Circ= dart_throw(int(dart_num))
        end_time = time()
        execution_time = end_time - start_time
        simulation_df.loc[int(dart_num)] = pd.Series({'Total_Darts': dart_num, 'Trial': run_num, 'Darts_in_Circ' : in_Circ, 'Execution_Time' : execution_time, 'Darts_per_second' : float(dart_num)/float(execution_time)})
    return simulation_df

def run_simulation_cluster(num_of_darts, run_num = 0):
    '''
    runs dart throwing simulation in a ipython clustering. The simulation calculates the execution time and darts per second thrown across the simulation length
    Simulation lenghth is determined by elements in the num_of_darts array, each entry specifies the length of the current simulation.
    input:
    number_of_darts -- an int array that specifices the number of darts to throw during a simulation.
    run_num -- if running multiple simulations of the same number of dart throws tags the trial number. 
    output:
    simulation_df -- returns a dataframe containing information for each simulation including the number of darts thrown in the simulation (Total_Darts), 
    the trial number (this is if you want to run through multiple simulations of the same length e.g. 5 simulations for 10, 100, 1000 etc darts thrown),
    the number of darts that landed in the circle (Darts_in_Circ), time to run the simulation of X number of dart throws (Execution_Time), and the 
    number of darts thrown per second during that simulation 
    '''
    simulation_df = pd.DataFrame(columns = ['Total_Darts', 'Trial', 'Darts_in_Circ', 'Execution_Time', 'Darts_per_second'], index = num_of_darts.astype(int))
    rc = parallel.Client()
    dview = rc[:]
    for dart_num in num_of_darts:
        start_time = time()
        result = dview.map_async(dart_throw, len(dview)*[int(dart_num)/len(dview)])
        in_Circ= sum(result.get())
        end_time = time()
        execution_time = end_time - start_time
        simulation_df.loc[int(dart_num)] = pd.Series({'Total_Darts': dart_num, 'Trial': run_num, 'Darts_in_Circ' : in_Circ, 'Execution_Time' : execution_time, 'Darts_per_second' : float(dart_num)/float(execution_time)})
    rc.close()
    return simulation_df

def run_simulation_multiproc(num_of_darts, run_num = 0):
    '''
    runs dart throwing simulation using multiprocesing. The simulation calculates the execution time and darts per second thrown across the simulation length
    Simulation lenghth is determined by elements in the num_of_darts array, each entry specifies the length of the current simulation.
    input:
    number_of_darts -- an int array that specifices the number of darts to throw during a simulation.
    run_num -- if running multiple simulations of the same number of dart throws tags the trial number. 
    output:
    simulation_df -- returns a dataframe containing information for each simulation including the number of darts thrown in the simulation (Total_Darts), 
    the trial number (this is if you want to run through multiple simulations of the same length e.g. 5 simulations for 10, 100, 1000 etc darts thrown),
    the number of darts that landed in the circle (Darts_in_Circ), time to run the simulation of X number of dart throws (Execution_Time), and the 
    number of darts thrown per second during that simulation 
    '''
    simulation_df = pd.DataFrame(columns = ['Total_Darts', 'Trial', 'Darts_in_Circ', 'Execution_Time', 'Darts_per_second'], index = num_of_darts.astype(int))
    numprocessors = cpu_count()
    pool = Pool(numprocessors)
    for dart_num in num_of_darts:
        start_time = time()
        #result = dview.map_async(dart_throw, len(dview)*[int(dart_num)/len(dview)])
        result = pool.map_async(dart_throw, 8*[int(dart_num)/8])
        in_Circ= sum(result.get())
        end_time = time()
        execution_time = end_time - start_time
        simulation_df.loc[int(dart_num)] = pd.Series({'Total_Darts': dart_num, 'Trial': run_num, 'Darts_in_Circ' : in_Circ, 'Execution_Time' : execution_time, 'Darts_per_second' : float(dart_num)/float(execution_time)})
    pool.close()
    return simulation_df
```

# Run Simulations

```python
#Run Serial simulation across 7 10 to 10^7 dart throws
num_of_darts = 10**sp.linspace(1, 7, 7)

serial_df = pd.DataFrame(columns = ['Total_Darts', 'Trial', 'Darts_in_Circ', 'Execution_Time', 'Darts_per_second'])

for n in range(5):
    serial_df = serial_df.merge(run_simulation(num_of_darts, run_num = n), how = "outer")

serial_df.head(7)
```

```python
#Create aggregate Serial data by averaging across trials
serial_Agg_df = pd.DataFrame(columns = ['Total_Darts',  'Execution_Time_Ave', 'Execution_Time_Std', 'Darts_per_second_Ave', 'Darts_per_second_Std'], index = num_of_darts.astype(int))
for dart_num in num_of_darts:
    serial_Agg_df.loc[int(dart_num)] = pd.Series({'Total_Darts': dart_num, 'Execution_Time_Ave': serial_df[serial_df['Total_Darts']==dart_num]['Execution_Time'].mean(),
                                                    'Execution_Time_Std': serial_df[serial_df['Total_Darts']==dart_num]['Execution_Time'].std(),
                                                    'Darts_per_second_Ave': serial_df[serial_df['Total_Darts']==dart_num]['Darts_per_second'].mean(),
                                                    'Darts_per_second_Std': serial_df[serial_df['Total_Darts']==dart_num]['Darts_per_second'].std()})
serial_Agg_df
```

```python
#Run Multiproc Simulations
multiProc_df = pd.DataFrame(columns = ['Total_Darts', 'Trial', 'Darts_in_Circ', 'Execution_Time', 'Darts_per_second'])

for n in range(5):
    multiProc_df = multiProc_df.merge(run_simulation_multiproc(num_of_darts, run_num = n), how = "outer")

multiProc_df.head(7)
```

```python
#Create aggregate multiproc data by averaging across trials
multProc_Agg_df = pd.DataFrame(columns = ['Total_Darts',  'Execution_Time_Ave', 'Execution_Time_Std', 'Darts_per_second_Ave', 'Darts_per_second_Std'], index = num_of_darts.astype(int))
for dart_num in num_of_darts:
    multProc_Agg_df.loc[int(dart_num)] = pd.Series({'Total_Darts': dart_num, 'Execution_Time_Ave': multiProc_df[multiProc_df['Total_Darts']==dart_num]['Execution_Time'].mean(),
                                                    'Execution_Time_Std': multiProc_df[multiProc_df['Total_Darts']==dart_num]['Execution_Time'].std(),
                                                    'Darts_per_second_Ave': multiProc_df[multiProc_df['Total_Darts']==dart_num]['Darts_per_second'].mean(),
                                                    'Darts_per_second_Std': multiProc_df[multiProc_df['Total_Darts']==dart_num]['Darts_per_second'].std()})
multProc_Agg_df
```

```python
#Run Cluster Simulations
cluster_df = pd.DataFrame(columns = ['Total_Darts', 'Trial', 'Darts_in_Circ', 'Execution_Time', 'Darts_per_second'])
for n in range(5):
    cluster_df = cluster_df.merge(run_simulation_cluster(num_of_darts, run_num=n), how="outer")
cluster_df.head(7)
```

```python
#Create aggregate cluster data by averaging across trials
cluster_Agg_df = pd.DataFrame(columns = ['Total_Darts',  'Execution_Time_Ave', 'Execution_Time_Std', 'Darts_per_second_Ave', 'Darts_per_second_Std'], index = num_of_darts.astype(int))
for dart_num in num_of_darts:
    cluster_Agg_df.loc[int(dart_num)] = pd.Series({'Total_Darts': dart_num, 'Execution_Time_Ave': cluster_df[cluster_df['Total_Darts']==dart_num]['Execution_Time'].mean(),
                                                    'Execution_Time_Std': cluster_df[cluster_df['Total_Darts']==dart_num]['Execution_Time'].std(),
                                                    'Darts_per_second_Ave': cluster_df[cluster_df['Total_Darts']==dart_num]['Darts_per_second'].mean(),
                                                    'Darts_per_second_Std': cluster_df[cluster_df['Total_Darts']==dart_num]['Darts_per_second'].std()})
cluster_Agg_df
```
