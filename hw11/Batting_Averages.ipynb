The hyperparameters alpha and beta should be chosen to reflect this belief. For this assignment, we will choose alpha and beta based on the league-wide averages from 2010. In 2010 the mean batting average was 0.255 and the variance between players was 0.0011. Choose alpha and beta to satisfy these prior beliefs!

```python
#Below used to calculate the beta distribution with a mean of 0.255 and variance of 0.0011
#Distribution Used for Priors
alpha = 43.5
beta = 127
x_theta = np.linspace(0, 1, 101) 
p_theta = stats.beta(alpha, beta).pdf(x_theta)
plot(x_theta, p_theta, linewidth=3)
print 'Distribution Mean: %f Distribution Var: %f' %(stats.beta(alpha, beta).mean(), stats.beta(alpha, beta).var())
```

## d) Compute the posterior mean and posterior 95% CI for each μi For how many of the 13 players does the full-season batting average fall within the 95% CI?

```python
# posterior estimation and inference for the batting average
statsDF = pd.DataFrame.from_dict(M.stats()).T
limitedDF = pd.DataFrame(columns = ['April_Avg', 'Posterior_Avg', 'CI_lower_bound', 'CI_upper_bound', 'True_Avg'], index = statsDF.index)

#generate new simplified dataframe with posterior average, 95% confidence intervals, True average for the season
for n in statsDF.index:
    curInd = int(n.replace('mu', ''))
    limitedDF.loc[n] = pd.Series({'April_Avg': april_data['AVG'][curInd], 'Posterior_Avg': statsDF.loc[n]['mean'], 'CI_lower_bound': statsDF.loc[n]['quantiles'][2.5], 
               'CI_upper_bound': statsDF.loc[n]['quantiles'][97.5], 'True_Avg': Full_data['AVG'][curInd]  })

#Add a column to check whether the true average was within the confidence interval generated by the posterior
limitedDF['Within_CI'] = (limitedDF['CI_lower_bound']<limitedDF['True_Avg']) & (limitedDF['True_Avg']<limitedDF['CI_upper_bound'])

#Column to illustrate the difference beteween the posterior average and the true full season average
limitedDF['Posterior_diff'] = np.abs(limitedDF['True_Avg']-limitedDF['Posterior_Avg'])

#column to illustrate the difference between the april average and the true full season average
limitedDF['Apr_diff'] = np.abs(limitedDF['True_Avg']-limitedDF['April_Avg'])

#column to illustrate whether the posterior average was a better prediction than the april average
limitedDF['Post_better'] = limitedDF['Posterior_diff']<limitedDF['Apr_diff']
HTML(limitedDF.to_html())
```

### 11 out of 13 players full season averages were within the confidence intervals
