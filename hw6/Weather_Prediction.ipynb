### Get Weather Data

```python
import urllib2
import re
from dateutil.relativedelta import relativedelta

def get_weather_fromURL(ICAO):
    '''
    Pulls weather data from weather underground
    for a given airport 1 month at a time and saves the results in a CSV file
    
    '''
    allStrings = []
    date = datetime.datetime(2008,1,1,12,4,5)
    allStrings = ['date,maxTemp,meanTemp,minTemp,maxDew,meanDew,' + 
                  'minDew,maxHum,meanHum,minHum,maxPress,meanPress,' +
                  'minPress,maxVis,meanVis,minVis,maxWind,'  +
                  'meanWind,maxGust,precip,clouds,events,windDirDeg']
    while date <= datetime.datetime.now(): 
       
        url = 'http://www.wunderground.com/history/airport/%s/%s/%s/%s/MonthlyHistory.html?format=1' %(ICAO, date.year, date.month, date.day)
        response = urllib2.urlopen(url)
        whetherText = response.read()
        monthIn = re.findall('(.*)<br />\n', whetherText)
        allStrings.extend(monthIn[1:])
        date += relativedelta(months=1)

    np.savetxt('./hw_6_data/%s_weather.csv' %ICAO, allStrings, delimiter=',', fmt="%s")

```

```python
engine.echo=False

#Compute Pairwise corrleations for maxTemp (T) and cloud cover (C)
one_dayTable_T = np.empty((len(ICAOs), len(ICAOs)), float)
three_dayTable_T = np.empty((len(ICAOs), len(ICAOs)), float)
seven_dayTable_T = np.empty((len(ICAOs), len(ICAOs)), float)
one_dayTable_C = np.empty((len(ICAOs), len(ICAOs)), float)
three_dayTable_C = np.empty((len(ICAOs), len(ICAOs)), float)
seven_dayTable_C = np.empty((len(ICAOs), len(ICAOs)), float)

for i, airport1 in enumerate(ICAOs):
    #Airport1, column -- current day
    for j, airport2 in enumerate(ICAOs):
        #Airport2, row -- future day
        if airport1 == airport2:
            #Set diagonal to zero to make it easier to get peaks
            one_dayTable_T[i, j] = 0
            three_dayTable_T[i, j] = 0
            seven_dayTable_T[i, j] = 0
            one_dayTable_C[i, j] = 0
            three_dayTable_C[i, j] = 0
            seven_dayTable_C[i, j] = 0
        else:
            #Query airport max Temp and cloud cover data
            s = sqlalchemy.sql.select([airports_weather.c.maxTemp, airports_weather.c.cloudCover], airports_weather.c.ICAO == airport1)
            airport1_weather = array(engine.execute(s).fetchall(), float)
            
            s = sqlalchemy.sql.select([airports_weather.c.maxTemp, airports_weather.c.cloudCover], airports_weather.c.ICAO == airport2)
            airport2_weather = array(engine.execute(s).fetchall(), float)
            
            
            one_dayTable_T[i, j] = sp.stats.pearsonr(airport1_weather[:-1, 0], airport2_weather[1:, 0])[0] 
            three_dayTable_T[i, j] = sp.stats.pearsonr(airport1_weather[:-3, 0], airport2_weather[3:, 0])[0]
            seven_dayTable_T[i, j] = sp.stats.pearsonr(airport1_weather[:-7, 0], airport2_weather[7:, 0])[0]
            
            one_dayTable_C[i, j] = sp.stats.pearsonr(airport1_weather[:-1, 1], airport2_weather[1:, 1])[0] 
            three_dayTable_C[i, j] = sp.stats.pearsonr(airport1_weather[:-3, 1], airport2_weather[3:, 1])[0]
            seven_dayTable_C[i, j] = sp.stats.pearsonr(airport1_weather[:-7, 1], airport2_weather[7:, 1])[0]
        
```

```python
#Find top 10 pairs
def get_topIndexes(corTable, numPull = 10):
    sortedRs = corTable.ravel()
    sortedRs = np.sort(sortedRs)[::-1]
    return np.where(corTable>=sortedRs[numPull-1])

from math import radians, cos, sin, asin, sqrt
def calc_distance(airport1, airport2, airportTable):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    lat1 = airportTable.loc[airport1, 'Lat']
    lon1 = airportTable.loc[airport1, 'Lon']
    lat2 = airportTable.loc[airport2, 'Lat']
    lon2 = airportTable.loc[airport2, 'Lon']
    
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    miles  = 6367 * c *.62137
    return miles 
```

```python
def plot_CorrxDist(corrTable, axesLabels, distTable, numPull=10, suptit=''):
    '''
    Plots the correlations strengths as a function of 1) distance and 2) lonitude
    input:
    corrTable -- Correlation matrix 
    axesLabels -- ICAO codes 
    distTable -- table that has information on longitude and latitude for the top 50 airports
    numPull -- how many pairs to plot
    '''
    #get top pairs
    sortedRs = corrTable.ravel()
    sortedRs = np.sort(sortedRs)[::-1]
    topPairsLocsCol, topPairsLocsRow = np.where(corrTable>=sortedRs[numPull-1])
    
    #initialize arrays
    PointNames = []
    distance = np.empty(np.shape(topPairsLocsCol.T), float)           #will contain distance using lat/long info
    pearR = np.empty(np.shape(topPairsLocsCol.T), float)              #pearson's R taken from correlation matrix
    longitudeDiff = np.empty(np.shape(topPairsLocsCol.T), float)      #longitude diference between two cities
    
    #Fill distance arrays as well as labels for the data points. Will label as the airport1 --> airport2 which in words is 
    #airport1 on given day predicts weather at airport2
    for i, pair in enumerate(zip(topPairsLocsCol, topPairsLocsRow)):
        PointNames.append(axesLabels[pair[0]] + '->' + axesLabels[pair[1]]) 
        distance[i] = calc_distance(axesLabels[pair[0]], axesLabels[pair[1]], distTable)
        pearR[i] = corrTable[pair[0], pair[1]]
        longitudeDiff[i] = distTable.loc[axesLabels[pair[0]], 'Lon'] - distTable.loc[axesLabels[pair[1]], 'Lon']
    
    #and for the plots...
    f, ax = plt.subplots(1, 2, figsize=(16, 6))
    
    #Plot correlations by lat/long distance
    ax[0].scatter(distance, pearR)
    ax[0].set_title('By Distance')
    ax[0].set_xlabel('Distance')
    ax[0].set_ylabel("Pearson's R")
    for label, x, y in zip(PointNames, distance, pearR):
        ax[0].annotate(
            label, 
            xy = (x, y), xytext = (-2, 2),
            textcoords = 'offset points', ha = 'right', va = 'bottom',
            fontsize='9')
    
    
    #Plot correlations by longitude distance
    ax[1].scatter(longitudeDiff, pearR)
    ax[1].set_title('By Longitude')
    ax[1].set_xlabel('Longitude Difference')
    ax[1].set_ylabel("Pearson's R")
    for label, x, y in zip(PointNames, longitudeDiff, pearR):
        ax[1].annotate(
            label, 
            xy = (x, y), xytext = (-2, 2),
            textcoords = 'offset points', ha = 'right', va = 'bottom',
            fontsize='9')
    f.suptitle(suptit)
```

### Trends?

In terms of max temperature and cloud cover as you'd predict you see greater correlations for 1 day diferences in predictions in cities that are closer together. For longer time periods, you see increases in the distances between the peak cities. There also seems to be a movement in one direction (as evident by the longitude plots) in that cities are west appear to do better at predicting cities that are more east. Also, temperature predictions seem to be much better across days than cloud cover.
