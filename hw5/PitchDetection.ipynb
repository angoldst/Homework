```python
def noteLookup(freq):
    '''
    Finds the closest note to the given frequency by chosing the note with the smallest distance between the fundamental frequency associated
    with the note and the inputed frequency. Note -- Frequency conversion table is stored in 'noteConversion.csv'
    '''
    noteKey = pd.DataFrame.from_csv('./noteConversion.csv', index_col = 0) #csv file with an array of freqency values paired with key/octaves
        
    notes = []
    true_freqs = []
    for f in freq:
        notes.append(noteKey.index[abs(noteKey['Freq'].values - f).argmin()]) ##Find closest note
        true_freqs.append(noteKey.loc[noteKey.index[abs(noteKey['Freq'].values - f).argmin()]][0]) ### find true/ideal frequency of that note
    
    
    notes = np.unique(np.array(notes))
    true_freqs = np.unique(np.array(true_freqs))

    print 'Peak based on highest power: \n%s' %noteKey.index[abs(noteKey['Freq'].values-freq[0]).argmin()]
    
    #test harmonics 
    harKey = np.zeros(len(true_freqs))
    for loc, i in enumerate(true_freqs):
        for j in true_freqs[loc:]:
            if (1.0*(j % i) / i > .98 or 1.0*(j % i) / i < .02) and i!=j: 
            #If the note frequency is a multiple of another frequency then consider the higher frequency as a harmonic overtone
                harKey[true_freqs==j]=1
    
    fundFreqs = np.unique(notes[harKey==0]) #Predicted non harmonics
    potHarms = np.unique(notes[harKey==1]) #predicted harmonics
    
    print 'Fundamental Frequency predictions: ' 
    for f in fundFreqs:
        print f
    
    print 'Harmonic Predictions: '
    for h in potHarms:
        print h

    return notes, true_freqs #retrun the index of the key vaule closest to the inputed freqency

def plot_audio_ampxtime(ax, y, rate):
    '''
    plots amplitude  against time data for an audio file
    '''
    time = arange(np.size(y)) / float(rate)
    plot_title = ("Audio Analysis")
    #fig = plt.figure(figsize=(7, 4.5))
    #ax1 = fig.add_subplot(1,1,1)
    ax.plot(time, y, color="red", linestyle="-")
    ax.set_xlabel("Time [s]")
    ax.set_ylabel("Amplitude")
    ax.set_xlim(min(time), max(time))
    ax.set_title(plot_title)
    #show()

def get_fft(ax, y, rate):
    '''
    Takes in a NxM dimensional array where N is the number of FFT chunks and M is the FFT length
    '''
    
    #calculate fft for each chunk
    F = 10*np.log10(np.abs(np.fft.rfft(y)))
    #print F.shape[1]
    #average across chunks 
    if len(F.shape)>1:
        F = np.mean(F, axis=0)
    #take first half of fft data and ignore redundancies
    #F = F[0:len(F)/2+1]
    
    #Get frequency bins
    num_bins = len(F)*.25
    xVals = np.arange(num_bins, dtype=np.float32)
    freq_Bins = xVals * (float(rate)/fft_length)

    #Plot Power Data 
    ax.plot(F[0:(len(F)-1)*.25])
    xTicks = np.round(ax.get_xticks()*float(rate)/fft_length).astype(int) #convert xticks into frequency 
    ax.xaxis.set_ticklabels(xTicks)
    ax.set_xlabel('Frequency (Hz)')  
    ax.set_ylabel('Power (DB)')
    
    return F, freq_Bins


def get_peaks(F, rate, fft_length):
    '''
    returns the frequencies with the highest power from the FFT
    Input --
    F -- FFT results from prior function
    rate == framerate 
    fft_length -- window length of fft
    
    '''
    #Calculate Frequency binds
    num_bins = len(F)-1
    xVals = np.arange(num_bins-1)
    freq_Bins = xVals * (float(rate)/fft_length)
    
    #Isolate peaks
    maxtab, mintab = peakdet(F, 10) #maxtab is a 2d array with the indicies/power
    
    #sort indecies by greatest power
    sorted_Inds = [x for (y, x) in sorted(zip(maxtab[:, 1], maxtab[:,0]), reverse=True)]
    
    #convert indices to int array to use as index
    ind = np.array(sorted_Inds, dtype=int)
    
    return freq_Bins[ind[0:3]]

import sys
from numpy import NaN, Inf, arange, isscalar, array

def peakdet(v, delta, x = None):
    """
    Converted from MATLAB script at http://billauer.co.il/peakdet.html
    
    Returns two arrays
    
    function [maxtab, mintab]=peakdet(v, delta, x)
    %PEAKDET Detect peaks in a vector
    %        [MAXTAB, MINTAB] = PEAKDET(V, DELTA) finds the local
    %        maxima and minima ("peaks") in the vector V.
    %        MAXTAB and MINTAB consists of two columns. Column 1
    %        contains indices in V, and column 2 the found values.
    %      
    %        With [MAXTAB, MINTAB] = PEAKDET(V, DELTA, X) the indices
    %        in MAXTAB and MINTAB are replaced with the corresponding
    %        X-values.
    %
    %        A point is considered a maximum peak if it has the maximal
    %        value, and was preceded (to the left) by a value lower by
    %        DELTA.
    
    % Eli Billauer, 3.4.05 (Explicitly not copyrighted).
    % This function is released to the public domain; Any use is allowed.
    
    """
    maxtab = []
    mintab = []
       
    if x is None:
        x = arange(len(v))
    
    v = asarray(v)
    
    if len(v) != len(x):
        sys.exit('Input vectors v and x must have same length')
    
    if not isscalar(delta):
        sys.exit('Input argument delta must be a scalar')
    
    if delta <= 0:
        sys.exit('Input argument delta must be positive')
    
    mn, mx = Inf, -Inf
    mnpos, mxpos = NaN, NaN
    
    lookformax = True
    
    for i in arange(len(v)):
        this = v[i]
        if this > mx:
            mx = this
            mxpos = x[i]
        if this < mn:
            mn = this
            mnpos = x[i]
        
        if lookformax:
            if this < mx-delta:
                maxtab.append((mxpos, mx))
                mn = this
                mnpos = x[i]
                lookformax = False
        else:
            if this > mn+delta:
                mintab.append((mnpos, mn))
                mx = this
                mxpos = x[i]
                lookformax = True

    return array(maxtab), array(mintab)
```
