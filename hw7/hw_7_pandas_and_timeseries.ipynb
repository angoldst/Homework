1) Make a DataFrame with one row per issue with the following columns extracted
from the issue data:

   ``ntitle, created_at, labels, closed_at, user, id``

Transform the user values to be simply the 'login' string, so that the user
column contains only string usernames.



```python
#load json file
json_data = open('./closed.json')
data = json.load(json_data, encoding='ascii')
json_data.close()
```

```python
#Convert json input into dataframe
allData_df = pd.DataFrame.from_dict(data)

#snag user dictionary and transform into another dataframe
user_dict = allData_df.user.to_dict()
user_DF = pd.DataFrame.from_dict(user_dict, orient='index')

#replace original user dictionary with string of login
allData_df['user'] = user_DF['login'].map(str)

#reduce large df to smaller one containing only relevent data
rel_data = allData_df[['title', 'created_at', 'labels', 'closed_at', 'user', 'id']]
rel_data

```

2) Remove duplicate rows by id from the DataFrame you just created using the id
column's duplicated method.



```python
rel_data = rel_data.ix[~rel_data.id.duplicated()]
rel_data
```

4) Convert the ``created_at`` and ``closed_at columns`` from string to datetime type.



```python
rel_data['created_at'] = pd.to_datetime(rel_data['created_at'])
rel_data['closed_at'] = pd.to_datetime(rel_data['closed_at'])

#set created at as index
rel_data = rel_data.set_index('created_at')
rel_data
```

5) Now construct appropriate time series and pandas functions to make the
following plots:

- Number of issues created by month

- Number of distinct users creating issues each month (hint: you can pass a
  function to resample's how argument, and there's nothing wrong with having
  string values in a TimeSeries)



```python
%matplotlib inline
#number of issues by month
n_issues_byMonth = rel_data['id'].resample('M', how=['count'])
n_issues_byMonth['count'].plot()

```

```python
def unique_users(in_ser):
    return len(in_ser.unique())
#number of distinct users
n_distict_users = rel_data['user'].resample('M', how=unique_users)
n_distict_users.plot()
```

6) Make a table and an accompanying plot illustrating:

- The mean number of days it took for issues to be closed by the month they
  were opened. In other words, for closed issues created in August 2012, how
  long were they open on average? (hint: use the ``total_seconds`` function on the
  timedelta objects computed when subtracting datetime objects). Also show the
  number of issues in each month in the table.



```python
#new table element showing time it took to close each issue
rel_data['time_to_close'] = pd.Series([(y - x).total_seconds() for x,y in zip(rel_data.index, rel_data['closed_at'])], index=rel_data.index)

to_close_byMonth = rel_data['time_to_close'].resample('M', how=['mean', 'count'])
to_close_byMonth['meanDays'] = to_close_byMonth['mean']/(60*60*24)
print to_close_byMonth

#plot the average time to close by month in seconds
to_close_byMonth['meanDays'].plot()
```

7) Make a DataFrame containing all the comments for all of the issues. You will
want to add an ``id`` attribute to each comment while doing so so that each row
contains a single comment and has the id of the issue it belongs to.

Convert the ``created`` column to datetime format; note you will need to multiply
the values (appropriately converted to integers) by 1000000 to get them in
nanoseconds and pass to to_datetime.



```python
#snag comments dictionary and transform into another dataframe
allData_DF_unique = allData_df.ix[~allData_df.id.duplicated()]

#get rid of non unique data
allData_DF_unique = allData_DF_unique.set_index('id') 

#convert comments entires to a dictionary
comments_dict = allData_DF_unique.comments.to_dict()
comments_df = pd.DataFrame()
issues= pd.DataFrame()
#Fill comments dataframe
for ID, issue in comments_dict.iteritems():
    for comment in issue:
            #print type(comment)
        comment.update({'id': ID})
        tempDF =  pd.DataFrame.from_dict(comment, orient='index').T
            #tempDF.index = ID
        issues = issues.append(tempDF)

#correct created to be used for datetime       
issues['created'] = issues['created'].map(int)*1000000

issues = issues.set_index('created')
issues.index = issues.index.to_datetime()
issues['created'] = issues.index
issues
```

8) For each month, compute a table summarizing the following for each month:

- Total number of issue comments
- The "chattiest" user (most number of comments)
- The percentage of total comments made by the chattiest users
- The number of distinct participants in the issue comments



```python
#sort dataframe
issues =  issues.sort_index()
```

```python
def most_Freq(x):
    u, indices = np.unique(np.array(x), return_inverse=True)
    try:
        return u[np.argmax(np.bincount(indices))]
    except:
        return 0

#percent of comments by most chatty users
def per_comm(x):
    u, indices = np.unique(np.array(x), return_inverse=True)
    try:
        author = u[np.argmax(np.bincount(indices))]
        #print author
        #print len(np.argwhere(x==author))
        return float(len(x[x==author]))/len(x)*100
    except:
        return 0    
    
#The number of distinct participants in the issue comments
def unique_authors(x):
    
    try:
        return len(x.unique())
    except:
        return 0

    
summary_table = issues.author.resample('M', how=['count', most_Freq, per_comm, unique_authors])
summary_table.columns = ['Num_Iss_Com', 'ChattiestUser', 'Chattiest_%_comm', 'num_participants']
summary_table
```

9) Create a helper ``labels`` table from the issues data with two columns: id and
label. If an issue has 3 elements in its 'labels' value, add 3 rows to the
table. If an issue does not have any labels, place a single row with None as
the label (hint: construct a list of tuples, then make the DataFrame).

```python
labels_listTemp = []
labels_data = []

#make higherarchical indecies for labels
for x, y in zip(allData_DF_unique.index, allData_DF_unique.labels):
    try:
        #if has 3 elements crreate tuples for color, name, url
        #will later convert to three extra rows
        y = dict(y[0])
        #print y['name']
        
        labels_listTemp.append(('color', x)) 
        labels_listTemp.append(('name', x))
        labels_listTemp.append(('url', x))
        
        labels_data.append([x, str(y['color'])])
        labels_data.append([x, str(y['name'])])
        labels_data.append([x, str(y['url'])])
    except:
        #otherwise have one row with none 
        labels_listTemp.append(('', x))
        labels_data.append([x, None])
        
```

```python
multInd = pd.MultiIndex.from_tuples(labels_listTemp, names=['label', 'ids'])
labels_table = pd.DataFrame(np.array(labels_data), index=multInd, columns=[['id', 'labels']])
labels_table.head(10)
```
