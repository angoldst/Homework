```python
#%%file FigureReplication.py

#Load Data
data = np.loadtxt('figData.txt', dtype=[("NegSD-NegSR", np.float32), 
    ("NeuSD-NeuSR", np.float32), ("AmbSD-AmbSR", np.float32)])

n = len(data) #number of subjects

#Calculate Means of data
NegMean = data["NegSD-NegSR"].mean()
NeuMean = data["NeuSD-NeuSR"].mean()
AmbMean = data["AmbSD-AmbSR"].mean()

#Calculate standard error of data
NegSem = data["NegSD-NegSR"].std()/np.sqrt(n)
NeuSem= data["NeuSD-NeuSR"].std()/np.sqrt(n)
AmbSem= data["AmbSD-AmbSR"].std()/np.sqrt(n)

N = 1
ind = np.arange(N)  # the x locations for the groups
width = 0.05         # the width of the bars
offset = .04        # space between the bars

f, ax = plt.subplots(figsize=(3, 4.5))
rect1 = ax.bar(ind, NegMean, width, color='r', yerr=NegSem, ecolor='k')
rect2 = ax.bar(ind+width+offset, NeuMean, width, color='y', yerr=NeuSem, ecolor='k')
rect3 = ax.bar(ind+2*(width+offset), AmbMean, width, color='w', yerr=AmbSem, ecolor='k')

# add some
ax.set_ylabel(r'$delta$')
ax.set_title('Scores by group and gender')
ax.set_xlim(0-offset, ind+width*3+offset*2+offset)
ax.set_ylim(0, 1.3)
ax.set_yticks((0, 1.3))
ax.set_xticks([])
#ax.set_xticklabels(('Sleep Deprived minus Sleep Rested'))
ax.set_xlabel('Sleep Deprived-Sleep Rested')
#ax.legend((rect1, rect2, rect3), ('NegCue', 'NeuCue', 'AmbCue'))
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.nxutils as nx
from matplotlib.patches import Rectangle
import matplotlib as mp
import sys
%matplotlib osx
class Annotate(object):
    def __init__(self, figure, axes, origData, cat, colors):
        self.figure = figure
        self.axes = axes
        self.ax = None
        self.mask = np.ones(len(origData), dtype=bool)
        self.colors = colors
        self.press = False
        self.exist = False
        self.colorDict = set_colorDict(cat, 
        self.x0 = None
        self.y0 = None
        self.x1 = None
        self.y1 = None
        self.curData = None
        self.origData = origData
        self.cat = cat
        self.axeID = figure.canvas.mpl_connect('axes_enter_event', self.get_Axis)
        self.keyID = figure.canvas.mpl_connect('key_press_event', self.on_key)
        self.mpressID = self.figure.canvas.mpl_connect('button_press_event', self.on_press)
        self.mreleaseID = self.figure.canvas.mpl_connect('button_release_event', self.on_release)
        #figure.canvas.mpl_connect('axes_leave_event', self.reset_Axis)
    
    def get_Axis(self, event):
        if self.press == False:
            self.ax = event.inaxes
            print 'entered axes %s %s' %(self.ax.rowNum, self.ax.colNum)
            #self.ax.figure.canvas.mpl_connect('motion_notify_event', self.on_motion)
        
    def clearRect(self):
        self.x0 = None
        self.y0 = None
        self.x1 = None
        self.y1 = None
        self.rect.remove()
        self.ax.figure.canvas.draw()
        self.mask = np.ones(len(self.origData), dtype = bool)
        #self.ax = None
        self.exist = False
        #self.ax.figure.canvas.mpl_disconnect(self.mpressID)
        #self.ax.figure.canvas.mpl_disconnect(self.mreleaseID)
        #self.mpressID = self.ax.figure.canvas.mpl_connect('button_press_event', self.on_press)
        #self.mreleaseID = self.ax.figure.canvas.mpl_connect('button_release_event', self.on_release)
        print 'cleared Rect'
        sys.stdout.flush()
    
    def on_key(self, event):
            if event.key == 'd' or event.key == 'D':
                    contains, attrd = self.rect.contains(event)
                    if not contains: return
                    self.clearRect()
                    re_makePlot(self.figure, self.axes, self.origData, self.mask, colors)
                    print 'pressed key'
                
    def on_press(self, event):
        if self.exist is True:
            self.clearRect()
        
        self.rect = Rectangle((0,0), 0, 0, fc='None', ec='k',alpha = .50)
        self.ax.add_patch(self.rect)
        self.press = True
        self.x0 = event.xdata
        self.y0 = event.ydata
        self.curData = get_Data(self.ax, self.origData)
        print 'pressed mouse'
        

    def on_release(self, event):
        self.press = False
        self.exist = True
    
        self.x1 = event.xdata
        self.y1 = event.ydata
        self.rect.set_width(self.x1 - self.x0)
        self.rect.set_height(self.y1 - self.y0)
        self.rect.set_xy((self.x0, self.y0))
        
        self.ax.figure.canvas.draw()
        #make_Op(self.ax, self.get_inRect(), self.colors)
        self.get_inRect()
        re_makePlot(self.figure, self.axes, self.origData, self.mask, self.colors)
        print 'released mouse'
        #print np.shape(self.get_inRect())
        
    def on_motion(self, event):
        #print 'on move'
        if self.press is False: return
        #if event.inaxes != self.rect.axes: return
        self.x1 = event.xdata
        self.y1 = event.ydata
        self.rect.set_width(self.x1 - self.x0)
        self.rect.set_height(self.y1 - self.y0)
        self.rect.set_xy((self.x0, self.y0))
        self.ax.figure.canvas.draw()
        

    def get_inRect(self):
        #x0 = self.rect.get_x()
        #x1 = self.rect.get_x() + self.rect.get_width()
        
        #y0 = self.rect.get_y()
        #y1 = self.rect.get_y() + self.rect.get_height()
        
        verts = [(self.x0, self.y0) , (self.x1, self.y0), (self.x0, self.y1), (self.x1, self.y1)]
        print verts
        #self.ax.set_clip_path(self.rect)
        print 'retested Rect'
        self.mask = nx.points_inside_poly(self.ax.transData.transform(self.curData), self.rect.get_verts())
        return nx.points_inside_poly(self.ax.transData.transform(self.curData), self.rect.get_verts())
        #return nx.points_inside_poly(self.curData, verts)



tests = [
    # (Type, Test)
    (np.int32, np.int32),
    (np.float32, np.float32),
]

def loadCSVFile(fname, in_dataType=[]):
    """ 
    loads text file and returns data array
    """
    if in_dataType==[]:
        types = []
        f = open(fname, 'rb')
        reader = csv.reader(f)
        names = reader.next()    
        
        dataIn = reader.next()
        for i in enumerate(dataIn):
            types.append(getType(i))
        
        in_dataType = np.dtype(dict(names=names, 
                               formats=types))
        
        
    dataOut = np.loadtxt(fname, in_dataType, skiprows=1, delimiter=',')
    return dataOut
    

def setColor(catArray, colormap = plt.cm.jet):
    #catArray = data["species"]
    #colormap = plt.cm.jet
    colorDtype =  np.dtype(tuple, (4,1))
    colorArray = np.zeros(np.shape(catArray), colorDtype)
    uniqCats = np.unique(catArray)
    inc = np.round(np.linspace(0, colormap.N/2, num=len(uniqCats))).astype(int)
    
    colorDict = dict()
    d = dict((key, value) for (key, value) in zip(uniqCats, inc))
    
    #colorArray =  tuple(colorArray)
    colorList = []
    for i in catArray:
        print "%s %s %s" %(i, d[i], colormap(d[i]))
        colorList.append(colormap(d[i]))
            
        return colorArray    

def removeScatters(ax):
     for intype in ax.get_children():
        if isinstance(intype, mp.collections.PathCollection):
            intype.remove()
            
def get_DataOld(ax):
    for intype in ax.get_children():
        if isinstance(intype, mp.collections.PathCollection):
            print type(intype)
            dataObj = intype
            print 'grabbed data from axis %s %s' %(ax.rowNum, ax.colNum)
            return dataObj.get_offsets()

def get_Data(ax, dataIn):
    col = ax.colNum
    row = ax.rowNum
    keys = dataIn.dtype.names
    print 'xData = %s', keys[col]
    print 'yData = %s', keys[row]
    dataOut = np.empty((len(dataIn), 2))
    dataOut[:, 0] = dataIn[keys[col]]
    dataOut[:, 1] = dataIn[keys[row]]
    return dataOut
    
    
def get_Scatter(ax):
    for intype in ax.get_children():
        if isinstance(intype, mp.collections.PathCollection):
            return intype

#def make_Op(ax, mask, colors):
#    """assumes there is only one scatter plot on given axis"""
    
#    scat = get_Scatter(ax)
#    data = get_Data(ax)
#    xData = data[:, 0]
#    print xData
#    yData = data[:, 1]
    #colors = scat.get_array()
    
#    print mask
    
#    print xData[mask]
    #ax.scatter(xData[mask],yData[mask], s=40, c=colors[mask], cmap = 'RdYlGn', alpha = 1)
    #ax.scatter(xData[-mask],yData[-mask], s=40, c=colors[-mask], cmap = 'RdYlGn', alpha = .05)
#    scat.remove()
    #ax.figure.canvas.draw()
    
    
def getType(value):
     for typ, test in tests:
         try:
             test(value)
             return typ
         except ValueError:
             continue
     # No match
     return "S10"

def re_makePlot(f, axes, dataIn, mask, colors):
    ''' re_makePlot(f, dataIn, mask, colors)
    f --      current figure
    dataIn -- original data
    mask --   mask derived from rectangle
    colors -- colors derived from category key
    '''
    
    keys = dataIn.dtype.names
    nRows = len(keys)
    nCols = len(keys)
    for c, cKey in enumerate(keys):
        for r, rKey in enumerate(keys):
            #oldScat = get_Scatter(axes[r, c])
            oldxLim = axes[r, c].get_xlim()
            oldyLim = axes[r, c].get_ylim()
            removeScatters(axes[r, c])
            xData = data[cKey]
            yData = data[rKey]
            axes[r, c].scatter(xData[mask],yData[mask], s=40, c=colors[mask], alpha = .4)
            
            if not mask.all():
                axes[r, c].scatter(xData[-mask],yData[-mask], s=40, c=colors[-mask], alpha = .02)
            axes[r, c].set_xlim(oldxLim)
            axes[r, c].set_ylim(oldyLim)
            #oldScat.remove()
            
    f.canvas.draw()
    
           

def makePlot(dataIn, colors):
    keys = dataIn.dtype.names
    
    if keys == None:
        f, ax = plt.subplots()
        scat = ax.scatter(data,data, s=40, c=colors, cmap = 'RdYlGn', alpha = .4)
    else:
        nRows = len(keys)
        nCols = len(keys)
        f, ax = plt.subplots(nRows, nCols, sharex='col', sharey='row')
        for c, cKey in enumerate(keys):
            for r, rKey in enumerate(keys):
                scat = ax[r,c].scatter(data[cKey],data[rKey], s=40, c=colors, alpha = .4)
                if r==c:
                    ax[r, c].text(0.1, 0.95, cKey,
            verticalalignment='top', horizontalalignment='left',
            transform=ax[r,c].transAxes,
            fontsize=10)
         
    return f, ax

```
