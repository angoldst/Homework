```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib 
```

```python
#%%file FigureReplication.py

#Load Data
data = np.loadtxt('figData.txt', dtype=[("NegSD-NegSR", np.float32), 
    ("NeuSD-NeuSR", np.float32), ("AmbSD-AmbSR", np.float32)])

n = len(data) #number of subjects

#Calculate Means of data
NegMean = data["NegSD-NegSR"].mean()
NeuMean = data["NeuSD-NeuSR"].mean()
AmbMean = data["AmbSD-AmbSR"].mean()

#Calculate standard error of data
NegSem = data["NegSD-NegSR"].std()/np.sqrt(n)
NeuSem= data["NeuSD-NeuSR"].std()/np.sqrt(n)
AmbSem= data["AmbSD-AmbSR"].std()/np.sqrt(n)

N = 1
ind = np.arange(N)  # the x locations for the groups
width = 0.05         # the width of the bars
offset = .04        # space between the bars

f, ax = plt.subplots(figsize=(3, 4.5))
rect1 = ax.bar(ind, NegMean, width, color='r', yerr=NegSem, ecolor='k')
rect2 = ax.bar(ind+width+offset, NeuMean, width, color='y', yerr=NeuSem, ecolor='k')
rect3 = ax.bar(ind+2*(width+offset), AmbMean, width, color='w', yerr=AmbSem, ecolor='k')

# add some
ax.set_ylabel(r'$delta$')
ax.set_title('Scores by group and gender')
ax.set_xlim(0-offset, ind+width*3+offset*2+offset)
ax.set_ylim(0, 1.3)
ax.set_yticks((0, 1.3))
ax.set_xticks([])
#ax.set_xticklabels(('Sleep Deprived minus Sleep Rested'))
ax.set_xlabel('Sleep Deprived-Sleep Rested')
#ax.legend((rect1, rect2, rect3), ('NegCue', 'NeuCue', 'AmbCue'))
```

```python
#load data skipping header rows
googleData = np.loadtxt('hw_3_data/google_data.txt', dtype=[("Date", np.int32), ("stock", np.float32)], skiprows=1)
tempData = np.loadtxt('hw_3_data/ny_temps.txt', dtype=[("Date", np.int32), ("temp", np.float32)], skiprows=1)
yahooData = np.loadtxt('hw_3_data/yahoo_data.txt', dtype=[("Date", np.int32), ("stock", np.float32)], skiprows=1)

allDates = np.concatenate((googleData["Date"], tempData["Date"], yahooData["Date"]))
allDates = np.unique(allDates)

f, ax1 = plt.subplots()

ln1 = ax1.plot(yahooData["Date"], yahooData["stock"], color='purple', label='Yahoo! Stock Value')
ln2 = ax1.plot(googleData["Date"], googleData["stock"], 'b', label = 'Google Stock Value')
ax1.set_ylabel('Value (Dollars)')

ax2 = ax1.twinx()
ln3 = ax2.plot(tempData["Date"], tempData["temp"], 'r--', label='NY Mon. High Temp')
ax2.set_ylabel(u'')

lns = ln1 + ln2 + ln3
labs = [l.get_label() for l in lns]
leg = ax1.legend(lns, labs, loc='center left', frameon=False)
plt.setp(leg)
ax2.set_ylim(-150, 100)

#plt.setp(line)
```
