```python
import numpy as np
import csv

#for testing incoming data types
tests = [  
    # (Type, Test)
    (np.int32, np.int32),
    (np.float32, np.float32),
]

def getType(value):
    '''
    used for testing input data from csv
    '''
    for typ, test in tests:
        try:
            test(value)
            return typ
        except ValueError:
            continue
     # No match
    return "S10"

def loadCSVFile(fname, in_dataType=[]):
    """ 
    loads text file and returns data array
    automaticaly sets data types using tests list above
    assumes there is a header row
    input:
        fname -- file to open
    returns:
        structred data array with headers
    """
    if in_dataType==[]:
        types = []
        f = open(fname, 'rb')
        reader = csv.reader(f)
        names = reader.next()
        print names
        
        dataIn = reader.next()
        for i in enumerate(dataIn):
            types.append(getType(i))
        
        in_dataType = np.dtype(dict(names=names, 
                               formats=types))
        
    try:    
        dataOut = np.loadtxt(fname, in_dataType, skiprows=1, delimiter=',')
        return dataOut
    except ValueError:
        print 'Check CSV file there should be no missing data' 
        
        
'''
plots figure 2 from Goldstein, AN, 2013 J Neuro paper
'''
dataA = loadCSVFile('figA.csv') #load Data from left portion of figure
n = len(dataA) #number of subjects

#Calculate Means and Error (SEM) of data
figAMeans = np.empty((3, 1))
figASEM = np.empty((3, 1))
for i, item in enumerate(dataA.dtype.names):
    figAMeans[i] = dataA[item].mean()
    figASEM[i] = dataA[item].std()/np.sqrt(n)
    

#Sets distance paramters for bar locations    
N = 1
ind = np.arange(N)  # the x locations for the groups
width = 0.05        # the width of the bars
offset = .01        # space between the bars
l_hang = 0.03       # distance from left y axis
r_hang = 0.06       # distance from right y axis

colors = [(1, 0, 0, 1), (1, 1, 0, 1), (1, 1, 1, 1)] #colors for bars


f, ax = plt.subplots(1, 2, figsize=(10, 8))

#generate bars for left part of figure
rect = []
for i in range(len(dataA.dtype.names)):
    rect.append(ax[0].bar(ind+(width+offset)*i, figAMeans[i], width, color=colors[i], yerr=([0], figASEM[i]), ecolor='k'))

#set figure 1 label/tick parameters
ax[0].set_ylabel('Sleep Deprived - Sleep Rested\n Contrast Estimates [au]', multialignment='center', size=15)
ax[0].set_xlim(0-l_hang, ind+width*3+offset*2+r_hang)
ax[0].spines['right'].set_color('none')
ax[0].spines['top'].set_color('none')
ax[0].set_ylim(0, 1.25)
ax[0].set_yticks((0, 1.25))
ax[0].set_xticks([])
ax[1].set_yticklabels([0, 1], size=12)

#make bars and *'s marking significance; ** = p<.01 * = p<.05 

#between Negative and Ambiguous 
ln1_0 = ind+width/2
ln1_1 = ind+width*2.5+offset*2 -.004
st_x = np.sqrt((ln1_1-ln1_0)**2)/2+ln1_0
ax[0].plot([ln1_0, ln1_1], [1.1, 1.1], color='k')  #plot sig line
ax[0].annotate('*', xy=(st_x, 1.10), size=20, weight='extra bold', multialignment='center') #plot sig star

#between Neutral and Ambiguous
ln2_0 = ind + width*1.5+offset
ln2_1 = ind + width*2.5 + 2*offset
st2_x = np.sqrt((ln2_1-ln2_0)**2)/2+ln2_0 -.004
ax[0].annotate('*', xy=(st2_x, 1.20), size=20, weight='extra bold', multialignment='center')
ax[0].plot([ln2_0, ln2_1], [1.2, 1.2], color='k')

#stars marking significance from zero
#negative star
dst_x0 = width/2 - .01 
ax[0].annotate('**', xy=(dst_x0, figAMeans[0]+figASEM[0]+.03), size=20, weight='extra bold', multialignment='center')
#neutral star
dst_x1 = width*1.5 - .01 + offset
ax[0].annotate('**', xy=(dst_x1, figAMeans[1]+figASEM[1]+.03), size=20, weight='extra bold', multialignment='center')

#generate legend for figure
leg = ax[0].legend(rect, ['Negative', 'Neutral', 'Ambiguous'], frameon=False, bbox_to_anchor=(0, 1.02, 1.5, .102), loc=3,
       ncol=3, mode="expand", borderaxespad=0., fontsize = 15)

#Figure part 2
dataB = loadCSVFile('figB.csv') #grab data for second part of figure


#Calculate Means and error of data
Conds = ['SD', 'SR']
Cues = ['Neg', 'Neu', 'Amb']

figBMeans = np.empty((3, 2))  
figBSEM = np.empty((3, 2))
for c, cond in enumerate(Conds):
    for cu, cue in enumerate(Cues):
        curName = cue + 'Cue' + cond
        figBMeans[cu, c] = dataB[curName].mean()
        figBSEM[cu, c] = dataB[curName].std()/np.sqrt(n)


#Sets distance paramters for bar locations            
N = 3
width = 0.01        # the width of the bars
offset = .02        # space between the bars
btw_offset = .002   # distance between pairs of data
l_hang = 0.006      # distance from left y axis
r_hang = 0.006      # distance from right y axis
ind = np.array([0 , 0+width+offset, 0+width*2+offset*2])   # the x locations for the groups


figBMeans += 1.5 # center data at -1.5 to match figure axes of published fig

#plot bars
for c in range(len(Conds)):
    ax[1].bar(ind+(width+btw_offset)*c, figBMeans[:, c], width, color=colors, yerr=([0, 0, 0],  figBSEM[:, c]), ecolor='k')

#set limits, tick and label parameters    
ax[1].set_xlim(0-l_hang, ind[-1]+width*2+r_hang)    
ax[1].set_ylim(0, 2)
ax[1].spines['right'].set_color('none')
ax[1].spines['top'].set_color('none')
ax[1].set_ylabel('Contrast Estimates [au]', multialignment='center', size=15)
ax[1].set_yticks([0, 2])
ax[1].set_yticklabels([-1.5, .5], size=12)

#make tick marks to match bar positioning
ticks = np.hstack((ind, ind+width+btw_offset)) 
ticks = np.sort(ticks)
ax[1].set_xticks(ticks)
plt.setp(ax[1].xaxis, ticks_position = 'none')
ax[1].set_xticklabels(['Deprived', 'Rested', 'Deprived', 'Rested', 'Deprived', 'Rested'], rotation = 50, size=14)

#make bars and *'s marking significance; ** = p<.01 * = p<.05 

#between NegativeSD and NegativeSR 
ln1_0 = ind[0] +width/2
ln1_1 = ind[0] + width*1.5 + btw_offset
st_0 = np.sqrt((ln1_1-ln1_0)**2)/2+ln1_0 - .004 
ax[1].plot([ln1_0, ln1_1], [figBMeans[0, 0]+figBSEM[0, 0]+.1, figBMeans[0, 0]+figBSEM[0, 0]+.1], color = 'k')
ax[1].annotate('**', xy=(st_0, figBMeans[0, 0]+figBSEM[0, 0]+.1), size=20, weight='extra bold', multialignment='center')

#between NeutralSD and NeutralSR
ln2_0 = ind[1] + width/2
ln2_1 = ind[1] + width*1.5 + btw_offset
st_1 = np.sqrt((ln2_1-ln2_0)**2)/2+ln2_0 - .002 
ax[1].plot([ln2_0, ln2_1], [figBMeans[1, 0]+figBSEM[1, 0]+.1, figBMeans[1, 0]+figBSEM[1, 0]+.1], color = 'k')
ax[1].annotate('*', xy=(st_1, figBMeans[1, 0]+figBSEM[1, 0]+.1), size=20, weight='extra bold', multialignment='center')

#between Ambiguous and Neutral 
ln3_0 =  ind[1] + width*1.5 + btw_offset
ln3_1 =  ind[2] + width*1.5 + btw_offset
st_2 = np.sqrt((ln3_1-ln3_0)**2)/2+ln3_0 - .002 
ax[1].plot([ln3_0, ln3_1], [figBMeans[1, 0]+figBSEM[1, 0]+.3, figBMeans[1, 0]+figBSEM[1, 0]+.3], color = 'k')
ax[1].annotate('*', xy=(st_2, figBMeans[1, 0]+figBSEM[1, 0]+.3), size=20, weight='extra bold', multialignment='center')
```
