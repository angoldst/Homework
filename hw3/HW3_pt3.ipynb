```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.nxutils as nx
from matplotlib.patches import Rectangle
import matplotlib as mp
import sys
%matplotlib osx


class Brush(object):
    def __init__(self, figure, axes, origData, catArray):
        '''
        initiates brush object
        figure --        initial figure pane passed in
        axes   --        initial list of axes passed in
        originalData --  plotting data only, does not include category data (i.e. float/ints only)
        catArray --      string array of category split, used to determine color settings 
        '''
        self.figure = figure           #base figure pane taken from input
        self.axes = axes               #base axes list taken from input
        self.ax = None                 #current working axis set in get_Axis() function
        self.mask = np.ones(len(origData), dtype=bool)       #mask adjusted by in_rect(), default all True 
        self.press = False             #tracks if mouse button is pressed down
        self.exist = False             #tracks if rectangle exists
        self.colorDict = makeColor_Dict(catArray)     #creates color dictionary based off of category axis
        self.x0 = None                 #initial x data point for generating rect, set when button is pressed in on_press()
        self.y0 = None                 #initial y data point for generating rect, set when button is pressed in on_press()
        self.x1 = None                 #end x data point for generating rect, set when button is released in on_relelease()
        self.y1 = None                 #end y data point for generating rect, set when button is released in on_relelease()
        self.curData = None            #current data from axes the rect is generated in, used to set mask with in_rect()
        self.origData = origData       #original data passed into the function, remains unchanged and to reset data after rect destroyed
        self.catArray = catArray       #string array that separates data into categories, used to set colors
        self.axeID = figure.canvas.mpl_connect('axes_enter_event', self.get_Axis)   #tracks current axis through get_Axis()
        self.keyID = figure.canvas.mpl_connect('key_press_event', self.on_key)      #tracks kill key to destroy rect in on_key()
        self.mpressID = self.figure.canvas.mpl_connect('button_press_event', self.on_press) #tracks mouse press
        self.mreleaseID = self.figure.canvas.mpl_connect('button_release_event', self.on_release) #tracks mouse release
        #self.motion = self.figure.canvas.mpl_connect('motion_notify_event', self.on_motion)
        #figure.canvas.mpl_connect('axes_leave_event', self.reset_Axis)
    
    def get_Axis(self, event):
        '''
        tracks mouse event when moving into new axes, then sets the current working axis (self.ax) to current window iff the mouse is not pressed
        '''
        if self.press == False: #check if mouse is pressed (i.e. generating rect)
            self.ax = event.inaxes     #gets current axes
            #print 'entered axes %s %s' %(self.ax.rowNum, self.ax.colNum)
            
        
    def clearRect(self):
        ''' 
        removes rectangle and information about rectangle, triggered after either a 'd' button press within brushed region
        as well as if new rectangle was generated with a mouse press
        '''
        self.x0 = None
        self.y0 = None
        self.x1 = None
        self.y1 = None
        self.rect.remove() #removes rectangle patch from axis
        self.ax.figure.canvas.draw() #refresh axis drawing
        self.mask = np.ones(len(self.origData), dtype = bool) #reset mask to include all data
        self.exist = False
        #print 'cleared Rect'
        sys.stdout.flush()
    
    def on_key(self, event):
        '''
        kill function for rectangle, only triggered if d or D were pressed and within rectangle space
        '''
        if event.key == 'd' or event.key == 'D':
                contains, attrd = self.rect.contains(event)
                if not contains: return
                self.clearRect()
                re_makePlot(self.figure, self.axes, self.origData, self.mask, self.catArray, self.colorDict) #reset opacity to default for all points
                #print 'pressed key'
                
    def on_press(self, event):
        ''' 
        triggered on mouse press begins generation of rectangle
        '''
        if self.exist is True: #delete old rectangle if one currently made
            self.clearRect()
        
        self.rect = Rectangle((0,0), 0, 0, fc='grey', ec='k',alpha = .50)
        self.ax.add_patch(self.rect)
        self.press = True
        self.x0 = event.xdata
        self.y0 = event.ydata
        self.curData = get_Data(self.ax, self.origData) #set working data to current axis 
        #print 'pressed mouse'
        

    def on_release(self, event):
        ''' 
        triggered on mouse release
        completes generation of rectangle and triggers replottimg
        '''
        
        
        if event.inaxes != self.ax:  #if outside bounds cancel rectangle and reset data
            self.clearRect()
            
        else:    
            self.x1 = event.xdata
            self.y1 = event.ydata    
            
            self.press = False
            self.exist = True
            
                
            self.rect.set_width(self.x1 - self.x0)
            self.rect.set_height(self.y1 - self.y0)
            self.rect.set_xy((self.x0, self.y0))
                
            self.ax.figure.canvas.draw()
            #make_Op(self.ax, self.get_inRect(), self.colors)
            self.get_inRect()
        re_makePlot(self.figure, self.axes, self.origData, self.mask, self.catArray, self.colorDict) #change opacity outside of rectangle
        #print 'released mouse'
        #print np.shape(self.get_inRect())
        
    def on_motion(self, event):
        '''
        triggered on mouse movement, shows rectangle generation with movement
        '''
        #print 'on move'
        if self.press is False: return #only trigger when mouse button is pressed
        #if event.inaxes != self.rect.axes: return
        self.x1 = event.xdata
        self.y1 = event.ydata
        self.rect.set_width(self.x1 - self.x0)
        self.rect.set_height(self.y1 - self.y0)
        self.rect.set_xy((self.x0, self.y0))
        self.figure.canvas.draw()
        

    def get_inRect(self):
        '''
        set and return mask of current data within rectangle mask
        '''
        
        
        self.mask = nx.points_inside_poly(self.ax.transData.transform(self.curData), self.rect.get_verts())
        return nx.points_inside_poly(self.ax.transData.transform(self.curData), self.rect.get_verts())
        #return nx.points_inside_poly(self.curData, verts)


#for testing incoming data types
tests = [  
    # (Type, Test)
    (np.int32, np.int32),
    (np.float32, np.float32),
]

def getType(value):
    '''
    used for testing input data from csv
    '''
    for typ, test in tests:
        try:
            test(value)
            return typ
        except ValueError:
            continue
     # No match
    return "S10"
            
def loadCSVFile(fname, in_dataType=[]):
    """ 
    loads text file and returns data array
    automaticaly sets data types using tests list above
    assumes there is a header row
    input:
        fname -- file to open
    returns:
        structred data array with headers
    """
    if in_dataType==[]:
        types = []
        f = open(fname, 'rb')
        reader = csv.reader(f)
        names = reader.next()
        print names
        
        dataIn = reader.next()
        for i in enumerate(dataIn):
            types.append(getType(i))
        
        in_dataType = np.dtype(dict(names=names, 
                               formats=types))
        
    try:    
        dataOut = np.loadtxt(fname, in_dataType, skiprows=1, delimiter=',')
        return dataOut
    except ValueError:
        print 'Check CSV file there should be no missing data' 
        


def makeColor_Dict(catArray, colormap = plt.cm.jet):
    '''
    create dictionary for color assigment using inputed category array and color map
    input:
        catArray --     string array length of the incoming data which will determine categories and coloring 
    returns:
        d -- a color index to category dictionary using a given color map
    
    '''
    uniqCats = np.unique(catArray)      #find unique categories 
    inc = np.round(np.linspace(0, colormap.N/2, num=len(uniqCats))).astype(int)     #get evenly distributed colors from color map based on num categories
    d = dict((key, value) for (key, value) in zip(uniqCats, inc))           #dictionary of color to category assignments
    return d

def setColor(catArray, d, colormap = plt.cm.jet):
    '''
    generate color list to be used for plotting functions, keeps categories same consistant colors
    
    inputs:
        catArray --      string array the length of working dataset 
        d --             color index to category dictionary will be used for color assignment
        colormap --      desired colormap current working colormap is jet
    returns:
        colorLIst --    a list of tuples to be used for color assignment in plotting functions derived from colorDict
    '''    
    colorList = []
    for i in catArray:
        #print "%s %s %s" %(i, d[i], colormap(d[i]))
        colorList.append(colormap(d[i])) #desired rgba tuple for given category
    return colorList 

def removeScatters(ax):
    ''' 
    clear current axis of scatter plots
    '''
    for intype in ax.get_children(): 
        if isinstance(intype, mp.collections.PathCollection): #finds scatter plots in current axis and removes them
            intype.remove()


def get_Data(ax, dataIn):
    ''' 
    returns data from the current axis
    input:
        ax -- current axis 
        dataIn -- original data with header information set as keys
    return:
        dataOut -- x and y values from current scatter plot
    '''
    col = ax.colNum
    row = ax.rowNum
    keys = dataIn.dtype.names                #get original names of data columns
    #print 'xData = %s', keys[col]
    #print 'yData = %s', keys[row]
    dataOut = np.empty((len(dataIn), 2))
    dataOut[:, 0] = dataIn[keys[col]]       #get x-values from current scatterplot
    dataOut[:, 1] = dataIn[keys[row]]       #get y-values from current scatterplot
    return dataOut

def re_makePlot(f, axes, dataIn, mask, catArray, colorDict):
    '''
    updates figure and opacity plots across all axes based on status of current rectangle
    inputs:
        f --      current figure
        dataIn -- original data
        mask --   mask derived from rectangle
        catArray -- category array the length of the same length as dataIn
        colorDict -- color index and category dictionary used to pick colors
    '''
    
    keys = dataIn.dtype.names
    nRows = len(keys)
    nCols = len(keys)
    for c, cKey in enumerate(keys):                    #redraw each axis 
        for r, rKey in enumerate(keys):
            #oldScat = get_Scatter(axes[r, c])
            oldxLim = axes[r, c].get_xlim()            #used to retain old x/ylims
            oldyLim = axes[r, c].get_ylim()
            removeScatters(axes[r, c])
            xData = data[cKey]
            yData = data[rKey]
            axes[r, c].scatter(xData[mask],yData[mask], s=40, c=setColor(catArray[mask], colorDict), alpha = .4)  #keep opacity of datapoints inside rect
            
            if not mask.all():
                axes[r, c].scatter(xData[-mask],yData[-mask], s=40, c=setColor(catArray[-mask], colorDict), alpha = .01) #reduce opactiy of datapoints outside rect
                
            axes[r, c].set_xlim(oldxLim)             #restore old x/ylims
            axes[r, c].set_ylim(oldyLim)
            
    f.canvas.draw()   #show changes
    
           
def makePlot(dataIn, catArray):
    ''' 
    makes initial plot from inported CSV
    inputs:
        dataIn -- structured input with keys from header information, either float or integer
        catArray -- string array determining way to split the data used to determine color 
    returns:
        f -- current figure frame
        ax -- matric of axes generated 
    '''    
    keys = dataIn.dtype.names
    colorDict = makeColor_Dict(catArray)
    if keys == None:
        print 'wrong type of datainput'
        #f, ax = plt.subplots()
        #scat = ax.scatter(data,data, s=40, c=setColor(catArray, colorDict), cmap = 'RdYlGn', alpha = .4)
    else:
        nRows = len(keys)
        nCols = len(keys)
        f, ax = plt.subplots(nRows, nCols, sharex='col', sharey='row')
        for c, cKey in enumerate(keys):
            for r, rKey in enumerate(keys):
                scat = ax[r,c].scatter(data[cKey],data[rKey], s=40, c=setColor(catArray, colorDict), alpha = .4)
                if r==c:
                    ax[r, c].text(0.1, 0.95, cKey,
            verticalalignment='top', horizontalalignment='left',
            transform=ax[r,c].transAxes,
            fontsize=10)
         
    return f, ax
```
