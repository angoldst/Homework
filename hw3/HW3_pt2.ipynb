```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#load data skipping header rows
googleData = np.loadtxt('hw_3_data/google_data.txt', dtype=[("Date", np.int32), ("stock", np.float32)], skiprows=1)
tempData = np.loadtxt('hw_3_data/ny_temps.txt', dtype=[("Date", np.int32), ("temp", np.float32)], skiprows=1)
yahooData = np.loadtxt('hw_3_data/yahoo_data.txt', dtype=[("Date", np.int32), ("stock", np.float32)], skiprows=1)

#generate figure and axes
f, ax1 = plt.subplots(figsize=(14, 12))

#make array of all dates to find min and max to generate xlim
allDates = np.concatenate((googleData["Date"], tempData["Date"], yahooData["Date"]))
allDates = np.unique(allDates)
ax1.set_xlabel('Date (MJD)', fontsize='20')

#plot yahoodata & google data against left y axis
ln1 = ax1.plot(yahooData["Date"], yahooData["stock"], color='purple', label='Yahoo! Stock Value', linewidth=2)
ln2 = ax1.plot(googleData["Date"], googleData["stock"], 'b', label = 'Google Stock Value', linewidth=2)

ax1.set_ylabel('Value (Dollars)', fontsize='20')
ax1.set_ylim(-20, 780)
plt.minorticks_on()

#generate second y-axis on right and plot temperature data
ax2 = ax1.twinx()
dashes = [8, 4] #specify dash width for line
ln3 = ax2.plot(tempData["Date"], tempData["temp"], 'r--', ms='40',  label='NY Mon. High Temp', linewidth=2, dashes=dashes) #
ax2.set_ylabel('Temperature (' + u'\xb0' + 'F)', fontsize='20')
ax2.set_ylim(-150, 100)
ax2.set_xlim(48800, 55600)
plt.minorticks_on()

#concatentate lines together to generate collated legend
lns = ln1 + ln2 + ln3
labs = [l.get_label() for l in lns]
leg = ax1.legend(lns, labs, loc='center left', frameon=False, prop={'size':16}, handlelength=3)

#generate title and lift it slightly off of axis
title = ax1.set_title('New York Temparature, Google, and Yahoo!', fontsize=40, fontname = 'times new roman', weight='extra bold')
title.set_y(1.04)
plt.subplots_adjust(top=0.86) 

#fine tune major and minor tick distances and thickness 
ax1.xaxis.set_tick_params(width=2, length=10, top=False, labelsize=14)
ax1.yaxis.set_tick_params(width=2, length=10, labelsize=14)
ax2.yaxis.set_tick_params(width=2, length=10, labelsize=14)

ax1.xaxis.set_tick_params(which='minor', width=1, length=5, top=False)
ax1.yaxis.set_tick_params(which='minor', width=1, length=5)
ax2.yaxis.set_tick_params(which='minor', width=1, length=5)

#make overall plot box thicker
plt.rc('axes', linewidth=3)

#save resulting figure as .png
f.savefig('StocksandTemp.png', bbox_inches=0)
```
