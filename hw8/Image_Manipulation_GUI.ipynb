```python
from Tkinter import *
import urllib2
import string 
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('TkAgg')
import sys
import urllib
import numpy as np
import scipy.ndimage as imgproc
import scipy.signal
from os.path import basename
from urlparse import urlsplit

from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2TkAgg
from matplotlib.figure import Figure
```

```python
def get_image(*args):
    '''
    Triggered when run query button fires. 
    Pulls query string from text entry (t) encodes in html and pulls the request from yahoo image search
    Subsequently downloads the first image and saves it in the current working directory as "dl_image.jpg"
    Updates the url text label field (urlText) as well as displays the loaded image in the display frame (im)
    '''
    #get query text
    get_info = {"p": t.get()} 
    
    #encode into html search parameters
    url_values = urllib.urlencode(get_info)  
    
    #run search
    url = "http://images.search.yahoo.com/search/images;_ylt=AwrTcX0L5GpSVI8AWiOZDWNH?%s&fr=&ei=utf-8&y=Search" %url_values

    # open the url and grab text
    response = urllib2.urlopen(url,data=url_values) 
    html = response.read()
    response.close()
    
    #find image url paths
    im_paths = re.findall("img .*?src='(.*?)'", html)
    
    #set imURL string variable to first image found
    imURL.set(im_paths[0])
    imgURL = im_paths[0]
    
    #save first image in search as a jpg
    imgData = urllib2.urlopen(imgURL).read()
    fileName = basename(urlsplit(imgURL)[2])
    output = open('dl_image.jpg','wb')
    output.write(imgData)
    output.close()
    
    #update current working image and redisplay root frame with updates 
    global sim
    sim = plt.imread('dl_image.jpg')
    ax.imshow(sim)
    im.show()
    root.update()
    
def blur_image(*args) :
    '''
    blurs image using 2d convolution. If the image is in color will display normally, if image is grey scale will show image
    using greyscale colormap
    
    '''
    global sim 
    tempSim = sim/255.   # normalise to 0-1, it's easier to work in float space

    # make some kernel
    t = 1 - np.abs(np.linspace(-1, 1, 21))
    kernel = t.reshape(21, 1) * t.reshape(1, 21)
    kernel /= kernel.sum()   # kernel should sum to 1!  :) 
    
    # convolve 2d the kernel with each channel
    if len(sim.shape)==3:
        r = scipy.signal.convolve2d(tempSim[:,:,0], kernel, mode='same')
        g = scipy.signal.convolve2d(tempSim[:,:,1], kernel, mode='same')
        b = scipy.signal.convolve2d(tempSim[:,:,2], kernel, mode='same')
        
        # stack the channels back into a 8-bit colour depth image and plot it
        tempSim = np.dstack([r, g, b])
    else:
        tempSim = scipy.signal.convolve2d(tempSim, kernel, mode='same')
    
    sim = (tempSim * 255).astype(np.uint8) 
    
    #display in correct color map
    if len(sim.shape)==3:
        ax.imshow(sim)
    else:
        ax.imshow(sim, cmap=plt.cm.Greys_r)
    
    #update image display to show alterations
    im.show()
    root.update()

def rotate(*args):
    '''
    rotates image 90 degrees
    '''
    global sim
    sim = imgproc.interpolation.rotate(sim, 90.0) #image rotation
    
    #display in correct color map
    if len(sim.shape)==3:
        ax.imshow(sim)
    else:
        ax.imshow(sim, cmap=plt.cm.Greys_r)
    
    #update display
    im.show()
    root.update()

def decolor(*args):
    '''
    if image is colored will turn it to grey scale. No change if image is already grey-scale
    '''
    global sim
    
    if len(sim.shape)==3:
        sim = np.mean(sim, axis=2)
        ax.imshow(sim, cmap = plt.cm.Greys_r)
        im.show()
        root.update()
        
def refresh(*args):
    '''
    reloads unmodified download image
    '''
    sim = plt.imread('dl_image.jpg')
    ax.imshow(sim)
    im.show()
    root.update()
```

```python
#Make input frame
root = Tk() ; root.title("Yahoo! Image Search")

#Set global variables that will be altered through gui
imURL = StringVar()         #string representing image URL
sim = np.empty((100, 100))  #array holder for downloaded image to complete manipulations

f, ax = plt.subplots(figsize=(3,6))

#Set up Top panel (Input) of GUI
inputFrame = LabelFrame(root, width=75, height=50, text="Input")
inputFrame.pack(side = TOP, fill=X)

iq = Label(inputFrame, textvariable="Query String:", text="Query String:")
t = Entry(inputFrame, text='Enter String', width=100)
ent = Button(inputFrame, text='Enter', command=get_image)    

iq.pack(side=LEFT, expand=1)
t.pack(side=LEFT, expand=1)
ent.pack(side=LEFT, expand=1)


#Make url text frame
urlTextFrame = LabelFrame(root, text="Image URL:")
urlTextFrame.pack(side = TOP, fill=X)

urlText = Label(urlTextFrame, textvariable=imURL)
urlText.pack(side=BOTTOM, fill=X, expand=1)

imURL.set("image URL")

#Make Figure Display Frame 
figFrame = LabelFrame(root, text="Image Display:", width=75, height=50)
figFrame.pack(side = TOP, fill=X)
im = FigureCanvasTkAgg(f, master=figFrame)
im.get_tk_widget().pack(side=TOP,fill=X,expand=1)

#sim = plt.imread('test.jpg')
#ax.imshow(sim)
#im.show()


#Make Bottom button frame
buttonFrame = LabelFrame(root, text="Image Manipulation Options:")
buttonFrame.pack(side=BOTTOM, fill=X)

r = Button(buttonFrame, text="Refresh", command=refresh)
filt = Button(buttonFrame, text="blur", command=blur_image)
rot = Button(buttonFrame, text="Rotate", command=rotate)
dec = Button(buttonFrame, text="Decolor", command=decolor)
q  = Button(buttonFrame, text="Quit",command=root.destroy)

r.pack(side=LEFT, expand=1)
filt.pack(side=LEFT, expand=1)
rot.pack(side=LEFT, expand=1)
dec.pack(side=LEFT, expand=1)
q.pack(side=LEFT,expand=1)

# run the main loop
root.mainloop()
```
